```python
from collections import deque

n, graph = None, None
d = [(-1, 0), (1, 0), (0, -1), (0, 1)]
UP, DOWN, RIGHT, LEFT = 0, 1, 2, 3


def user_input():
    global n, graph
    n = int(input())
    graph = [list(map(int, input().split())) for _ in range(n)] if n > 1 else [int(input())]


def move(graph, d):
    max_num = 0
    new_graph = []
    if d == UP:
        for j in range(n):  # j열 처리
            col = [graph[c][j] for c in range(n) if graph[c][j] != 0]  # 열에서 0 제거
            new_col = []
            for num in col:
                if new_col and not new_col[-1][1] and new_col[-1][0] == num:  # 마지막 숫자가 사용되지 않았고, 같다면
                    last_num = new_col.pop()[0]
                    new_col.append([last_num * 2, True])  # 2배 하고, 사용 처리
                else:
                    new_col.append([num, False])
            new_col = [item[0] for item in new_col]  # 숫자만 저장
            new_col += [0] * (n - len(new_col))  # 뒤에 0 추가
            max_num = max(max_num, *new_col)  # 최댓값 갱신
            new_graph.append(tuple(new_col))
        new_graph = list(map(tuple, zip(*new_graph)))  # 뒤집기
    elif d == DOWN:
        for j in range(n):
            col = [graph[c][j] for c in range(n) if graph[c][j] != 0]  # 열에서 0 제거
            new_col = []
            for num in col[::-1]:  # 역순으로 탐색
                if new_col and not new_col[-1][1] and new_col[-1][0] == num:  # 마지막 숫자가 사용되지 않았고, 같다면
                    last_num = new_col.pop()[0]
                    new_col.append([last_num * 2, True])  # 2배 하고, 사용 처리
                else:
                    new_col.append([num, False])
            new_col = [item[0] for item in new_col][::-1]  # 숫자만 저장 후 뒤집기
            new_col = [0] * (n - len(new_col)) + new_col  # 앞에 0 추가
            max_num = max(max_num, *new_col)  # 최댓값 갱신
            new_graph.append(tuple(new_col))
        new_graph = list(map(tuple, zip(*new_graph)))  # 뒤집기
    elif d == RIGHT:
        for i in range(n):  # 행 탐색
            row = [graph[i][c] for c in range(n) if graph[i][c] != 0]  # 행에서 0 제거
            new_row = []
            for num in row[::-1]:  # 역순으로 탐색
                if new_row and not new_row[-1][1] and new_row[-1][0] == num:  # 마지막 수자가 사용되지 않았고, 같다면
                    last_num = new_row.pop()[0]
                    new_row.append([last_num * 2, True])  # 2배 하고, 사용 처리
                else:
                    new_row.append([num, False])
            new_row = [item[0] for item in new_row][::-1]  # 숫자만 저장 후 뒤집기
            new_row = [0] * (n - len(new_row)) + new_row
            max_num = max(max_num, *new_row)  # 최댓값 갱신
            new_graph.append(tuple(new_row))

    elif d == LEFT:
        for i in range(n):  # 행 탐색
            row = [graph[i][c] for c in range(n) if graph[i][c] != 0]  # 행에서 0 제거
            new_row = []
            for num in row:  # 탐색
                if new_row and not new_row[-1][1] and new_row[-1][0] == num:  # 마지막 수자가 사용되지 않았고, 같다면
                    last_num = new_row.pop()[0]
                    new_row.append([last_num * 2, True])  # 2배 하고, 사용 처리
                else:
                    new_row.append([num, False])
            new_row = [item[0] for item in new_row]  # 숫자만 사용
            new_row += [0] * (n - len(new_row))
            max_num = max(max_num, *new_row)  # 최댓값 갱신
            new_graph.append(tuple(new_row))

    return tuple(new_graph), max_num


def bfs():
    if n == 1:
        print(graph[0])
        return
    q = deque()
    answer = 0
    q.append((tuple(map(tuple, graph)), 0, 0))
    cash = set()

    while q:
        board, max_num, count = q.popleft()
        cash.add(board)
        if max_num > answer: answer = max_num
        for i in range(4):
            new_board, new_max_num = move(board, i)
            if count < 5 and new_board not in cash:
                q.append((new_board, new_max_num, count + 1))
    print(answer)


# --- TEST
def test_input():
    global n, graph
    n = 1
    # graph = [[2, 2, 2], [4, 4, 4], [8, 8, 8]]
    # graph = [[0, 0, 0], [2, 4, 2], [2, 2, 2]]
    # graph = [[2, 0, 0, 0, 0], [2, 0, 0, 0, 0], [4, 0, 0, 0, 0], [2, 0, 0, 0, 0], [2, 0, 0, 0, 0]]
    graph = []


def test_move():
    new_graph, max_num = move(graph, UP)
    print('new graph: ', new_graph)
    print('max_num: ', max_num)


# --- exec
# test_input()
user_input()
# test_move()

bfs()
```
