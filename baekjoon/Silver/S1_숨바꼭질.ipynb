## 시간 초과

```python
from collections import deque
inf = 10e+9

n, k = map(int, input().split())
n, k = min(n, k), max(n, k)

d = [inf] * (k*10)
d[n] = 0

queue = deque([n])
while queue:
    
    
    print('queue = ', queue, end= '  ')
    print("*" * 30)
    for i, dd in enumerate(d):
        print(dd, end=' ') if dd != inf else print('x', end=' ')
        if not (i+1) % 10:
            print()
    print('\n\n')
    
    num = queue.popleft()
    d[num] = min(d[num], num)
    if num == k:
        print('answer: ', d[num])
        break
    if num in range(1, k*2+2):
        if d[num-1] > d[num]+1:
            print(f'{num-1} 추가')
            d[num-1] = d[num]+1
            queue.append(num-1)
        if d[num+1] > d[num]+1:
            print(f'{num+1} 추가')
            d[num+1] = d[num]+1
            queue.append(num+1)
        if not num%2 and d[num//2] > d[num]+1:
            print(f'{num//2} 추가')
            d[num//2] = d[num]+1
            queue.append(num//2)
        if d[num*2] > d[num]+1:
            print(f'{num*2} 추가')
            d[num*2] = d[num]+1
            queue.append(num*2)
    print("\n\n\n")
print('answer: ', d[k])
```

## 메모리 초과

```python
# BFS
from collections import deque

n, k = map(int, input().split())
queue = deque([(0, n)])
visited = [n]

while queue:
    answer, num = queue.popleft()
    if num == k:
        print(answer)
        break
    
    if num > 0 and num-1 not in visited:
        visited.append(num-1)
        queue.append((answer+1, num-1))
    if num+1 not in visited:
        visited.append(num+1)
        queue.append((answer+1, num+1))
    if num*2 not in visited:
        visited.append(num*2)
        queue.append((answer+1, num*2))
```

## 정답

```python
# BFS
from collections import deque

n, k = map(int, input().split())
queue = deque([n])
d = [0] * 100001

while queue:
    num = queue.popleft()
    if num == k:
        print(d[k])
        break
    for nx in (num-1, num+1, num*2):
        if nx in range(0, 100001) and not d[nx]:
            d[nx] = d[num]+1
            queue.append(nx)
```
