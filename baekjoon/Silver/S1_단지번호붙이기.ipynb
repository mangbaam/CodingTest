```python
from collections import deque as queue

n = int()
arr = list()
visited = list()
total, size = 0, list()

dx = [-1, 0, 1, 0] # ↑ ← ↓ →
dy = [0, -1, 0, 1] # ↑ ← ↓ →

def user_input():
    global n, arr, visited

    n = int(input())
    arr = [list(map(int, list(input()))) for _ in range(n)]
    visited = [[False] * n for _ in range(n)]

def bfs(x, y):
    global n, arr, visited, size

    count = 0
    visited[x][y] = True
    q = queue([(x, y)])
    while q:
        a, b = q.popleft()
        print(f'arr[{a}][{b}] 방문!')
        count += 1
        for i in range(4):
            nx, ny = a + dx[i], b + dy[i]
            if nx in range(n) and ny in range(n) and arr[nx][ny] and not visited[nx][ny]:
                print(f'\tarr[{nx}][{ny}] enqueue!')
                visited[nx][ny] = True
                q.append((nx, ny))
        print(f'> 현재 큐: {q}\n')
    return count

def solution():
    global n, arr, total

    user_input()

    for x in range(n):
        for y in range(n):
            if not arr[x][y] or visited[x][y]: continue
            total += 1
            size.append(bfs(x, y))

solution()

print(total)
for num in size:
    print(num)
```
