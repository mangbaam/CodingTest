```python
def check(num1, b, num2):
    return num1 < num2 if b == '<' else num1 > num2

def find(count, s):
    if count == n + 1:
        answer.append(s)
        return
    for i in range(10):
        if not visited[i]:
            if count == 0 or check(s[-1], bracket[count - 1], str(i)):
                visited[i] = True
                find(count + 1, s + str(i))
                visited[i] = False

visited = [False] * 10
n = int(input())
bracket = input().split(' ')
answer = []

find(0, "")
print(answer[-1])
print(answer[0])
```
