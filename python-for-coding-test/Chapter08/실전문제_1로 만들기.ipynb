```python
x = int(input())
d = [0] * 30001

for i in range(2, x+1):
    li = []
    if i % 5 == 0:
        li.append(d[i//5])
    if i % 3 == 0:
        li.append(d[i//3])
    if i % 2 == 0:
        li.append(d[i//2])
    li.append(d[i-1])
    d[i] = min(li) + 1
print(d)
print(d[x])
```
